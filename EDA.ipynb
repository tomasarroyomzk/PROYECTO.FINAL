```python


genre_cols = df.columns[2:23]
genre_counts = df[genre_cols].sum().sort_values(ascending=False)

plt.figure(figsize=(15, 7))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.xticks(rotation=45)
plt.title("DistribuciÃ³n de GÃ©neros")
plt.ylabel("Cantidad")
plt.show()
```

```python
sns.countplot(data=df, x='year')
plt.title("NÃºmero de ReseÃ±as por AÃ±o")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df, x='recommend')
plt.title("DistribuciÃ³n de ReseÃ±as Recomendadas vs No Recomendadas")
plt.show()
```
